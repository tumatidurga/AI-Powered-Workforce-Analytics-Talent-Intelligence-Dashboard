**Prompt Used :** Create a SQL database in Google Colab using SQLite. Create a table named Employees with workforce details and insert 20 employee records into it. Display all the records to verify the data has been inserted successfully.Display all the records to verify the data has been inserted successfully.And perform operations like alter, delete , joins, drop.

```python
# Import SQLite
import sqlite3

# Create a database
conn = sqlite3.connect("company.db")

# Create cursor
cursor = conn.cursor()

# Create Employees table
cursor.execute("""
CREATE TABLE Employees (
    EmpID INTEGER PRIMARY KEY,
    Name TEXT,
    Age INTEGER,
    Gender TEXT,
    Department TEXT,
    Salary REAL
)
""")

print("Employees table created successfully.")
```

**Insert 10 Employee Records**

```python
employees = [
    (101, "Rahul", 25, "Male", "HR", 35000),
    (102, "Priya", 28, "Female", "IT", 50000),
    (103, "Amit", 30, "Male", "Finance", 45000),
    (104, "Sneha", 27, "Female", "Marketing", 40000),
    (105, "Kiran", 35, "Male", "IT", 65000),
    (106, "Anjali", 26, "Female", "HR", 38000),
    (107, "Vijay", 32, "Male", "Sales", 48000),
    (108, "Divya", 29, "Female", "Finance", 52000),
    (109, "Arjun", 31, "Male", "Marketing", 55000),
    (110, "Pooja", 24, "Female", "Sales", 36000)
]

cursor.executemany("INSERT INTO Employees VALUES (?, ?, ?, ?, ?, ?)", employees)

conn.commit()

print("10 records inserted successfully.")
```

**Display All Records**

```python
cursor.execute("SELECT * FROM Employees")

rows = cursor.fetchall()

print("Employee Records\n")

for row in rows:
    print(row)
```

**ALTER Operation**

```python
cursor.execute("""
ALTER TABLE Employees
ADD COLUMN Email TEXT
""")

conn.commit()

print("Column Email added successfully.")
```

**Update Email values**

```python
cursor.execute("""
UPDATE Employees
SET Email='rahul@gmail.com'
WHERE EmpID=101
""")

cursor.execute("""
UPDATE Employees
SET Email='priya@gmail.com'
WHERE EmpID=102
""")

conn.commit()
```

**Display table**

```python
cursor.execute("SELECT * FROM Employees")

for row in cursor.fetchall():
    print(row)
```

**DELETE Operation**

```python
cursor.execute("""
DELETE FROM Employees
WHERE EmpID=110
""")

conn.commit()

print("Record deleted.")
```

**JOIN Operation**

```python
cursor.execute("""
CREATE TABLE Projects(
    ProjectID INTEGER PRIMARY KEY,
    EmpID INTEGER,
    ProjectName TEXT
)
""")

projects = [
    (1,101,"Payroll System"),
    (2,102,"Website"),
    (3,103,"Banking App"),
    (4,105,"AI Chatbot"),
    (5,108,"ERP System")
]

cursor.executemany("INSERT INTO Projects VALUES(?,?,?)", projects)

conn.commit()

print("Projects table created.")
```

**INNER JOIN**

```python
cursor.execute("""
SELECT Employees.EmpID,
Employees.Name,
Employees.Department,
Projects.ProjectName
FROM Employees
INNER JOIN Projects
ON Employees.EmpID=Projects.EmpID
""")

rows = cursor.fetchall()

print("INNER JOIN Result\n")

for row in rows:
    print(row)
```

**DROP Operation**

```python
cursor.execute("""
DROP TABLE Projects
""")

conn.commit()

print("Projects table dropped successfully.")
```

**Close Database**

```python
conn.close()

print("Database closed.")
```
